5. Используйте другие метрики качества регрессионной модели в дополнение к коэффициенту детерминации.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score
```

```python
#оценка производительности модели на данных
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)   # разбивает на тестовые и обучающие данные
slr = LinearRegression()

slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)   # предсказание обученных данных
y_test_pred = slr.predict(X_test)    # предсказание тестовых данных
```

```python
plt.scatter(y_train_pred,  y_train_pred - y_train,
            c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred,  y_test_pred - y_test,
            c='lightgreen', marker='s', label='Test data')
plt.legend(loc = 'upper right')
plt.title('оценка производительности модели на данных')
```

```python
print('MSE train: {:.3f}, test: {:.3f}'.format(           # среднеквадратичная ошибка
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))    # коэф детерминации
#MSE насколько сильно модель ошибается в квадратных единицах
#как хорошо модель обобщает данные
#R  ближе к 1 означает лучшее соответствие
```

1. Применение любых моделей машинного обучения начинается с подготовки данных в необходимом формате. Для этого очень удобными для нас будут библиотеки numpy и pandas. Они практически всегда используются совместно с библиотекой sckikit-learn и другими инструментами машинного обучения. В первую очередь мы будем использовать numpy для создания массивов и операций с векторами и матрицами. Pandas нам понадобится для работы с табличными структурами - датасетами.

Если вы хотите самостоятельно задать в явном виде данные обучающей выборки, то нет ничего лучше использования обычных массивов ndarray. Обычно в одном массиве хранятся значения атрибутов - x, а в другом - значения целевой переменной - y.
