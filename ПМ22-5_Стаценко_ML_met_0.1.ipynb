##17.

image.png

```python
room_count = pd.value_counts(df['Rooms']).to_dict()
labels = room_count.keys()
sizes = room_count.values()
room_count
```

```python
explode = [0.1 if i == 0 else 0 for i in range(len(labels))]

# отрисовка
plt.figure(figsize=(10, 10))
plt.title('Количество квартир в процентах в зависимости от количества комнат')
colors = sns.color_palette('pastel')[0:len(labels)]

plt.pie(sizes, labels=labels, colors=colors, startangle=100, labeldistance=1.1, explode=explode)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), labels = ['{0} - {1} %'.format(i,j) for i,j in zip(labels, [i/100 for i in sizes])])

plt.show()
plt.clf()
```

##18.

image.png

```python
plt.figure(figsize=(10, 6))
plt.title('Гистограмма по целевой переменной Price')
sns.histplot(data=df, x="Price")
plt.axvline(x=np.median(df["Price"]), color='red', linestyle='--', label=f'цена наибольшего кол-во квартир: {np.median(df["Price"])}')
plt.legend()
plt.show()
```

##19.

image.png

```python
plt.figure(figsize=(15, 10))
plt.title('Диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от Price')

plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='Rooms', y='Price')

plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='Square', y='Price')

plt.subplot(2, 2, 3)
sns.scatterplot(data=df, x='HouseFloor', y='Price')

plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='HouseYear', y='Price')

plt.show()
# разброс точек близок к линейной функции на диаграмме рассеятния для признака HouseYear (x=0)
```

##20.

image.png

```python
plt.figure(figsize=(10, 5))
plt.title('Ядерная оценка плотности для Price')
sns.kdeplot(data=df, x='Price', fill=True)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10, 6))
sns.jointplot(data=df, x='HouseFloor', y='Price', kind='kde', fill=True)
plt.suptitle('Двумерная ядерная оценка плотности для Price и HouseFloor', y=1.02)
plt.show()
plt.clf()
```

image.png

```python
g = sns.PairGrid(df[['Rooms', 'Square', 'HouseFloor', 'Price']])
g.map_diag(sns.histplot, kde_kws={'color': 'k'}) # гистограммы
g.map_lower(sns.kdeplot, cmap='Blues_d') # ядерные оценки плотности
g.map_upper(sns.scatterplot) # диаграммы рассеяния
g.fig.suptitle('Попарные отношения признаков Rooms, Square, HouseFloor, Price', y=1.02)
plt.show()
# HouseFloor, HouseYear
```

Выводы:  
- чем больше комнат и чем больше площадь, тем выше зп
- многокомнатные квартир, где комнат больше 5 почти не встречаются на высоких этажах (30-45)

```python
df_temp = df[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']]
correlation_matrix = df_temp.corr() # матрица корр.

# Создание тепловой карты
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Тепловая карта матрицы корреляции')
plt.show()
```
